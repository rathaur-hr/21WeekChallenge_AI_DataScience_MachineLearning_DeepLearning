#Importing Seaborn

```python
import seaborn as sns
```

#Importing Matplotlib, Numpy & Pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#Seaborn Builtin Data Sets

**Total Bill vs Total Tip Dataset**

```python


tips = sns.load_dataset('tips')
tips.head()
```

```python
#Setting a theme for plots

sns.set_theme() #Default theme set using this
```

```python
#Visualizing the data

sns.relplot(data=tips, x='total_bill', y='tip',col='time',hue='smoker',style='smoker',size='size')
```

**Iris Dataset**

```python
#Loading iris dataset

iris = sns.load_dataset("iris")
iris.head()
```

```python
#Plotting iris Data set usning scatter plot
#parameter used = sepallength and petallength

sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
#Plotting iris Data set usning scatter plot
#parameter used = sepallength and petalwidth

sns.scatterplot(x='sepal_length',y='petal_width',hue='species',data=iris)
```

**Titanic Dataset**


```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#Count Plot using titanic Dataset

sns.countplot(x='class',data=titanic)
```

```python
sns.countplot(x='survived',data=titanic)
```

```python
#Bar Chart

sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
#Distribution Plot

sns.distplot(titanic['fare'])
```

**Heat Map Plot**

```python
diabetes = pd.read_csv('/diabetes.csv')
correlation = diabetes.corr()
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, cmap = 'Blues')
plt.figure()
```
