#Importing Matplotlib Library

```python
import matplotlib.pyplot as plt
```

#Importing Numpy and Pandas

```python
import numpy as np
import pandas as pd
```

#Matplotlib Program

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print("Value of X")
print(x)
print("Value of Y")
print(y)
print("Value of Z")
print(z)
```

```python
#Plotting Above Values as Graph using Matplotlib (Sine Wave)
plt.plot(x,y)
plt.show()
```

```python
#Plotting Above Values as Graph using Matplotlib (Cosine Wave)
plt.plot(x,z)
plt.show()
```

```python
#Adding Title X-Axis and Y-Axis Labels

plt.plot(x,y)
plt.xlabel("Angle")
plt.ylabel("Sine Value")
plt.title("Sine Wave")
print(plt.show())
```

```python
#Plotting Parabolla

x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

#Plotting Using Different Symbol


```python
#Plotting Using + Symbol

plt.plot(x,y,"b+")
plt.show()
```

```python
#Plotting Using * Symbol

plt.plot(x,y,"b*")
plt.show()
```

```python
x = np.linspace(-5,+5,50)
plt.plot(x, np.sin(x),"g-")
plt.plot(x, np.cos(x),"b--")
plt.show()
```

```python
#Bar Plots

figure = plt.figure()
ax = figure.add_axes([0,0,1,1])
languages = ['Hindi','Maithili', 'Bhojpuri', 'English','Latin']
peoples = [51, 75, 46, 25, 31]
ax.bar(languages,peoples)
plt.xlabel('Languages')
plt.ylabel('No Of Peoples')
plt.show()
```

```python
#Pie Chart

fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['Hindi','Maithili', 'Bhojpuri', 'English','Latin']
peoples = [51, 75, 46, 25, 31]
ax.pie(peoples, labels=languages, autopct='%1.1f%%')
plt.show()
```

```python
#Scatter Plot

x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
#3D Scatter Plot

fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap="Blues")
plt.show()
```
